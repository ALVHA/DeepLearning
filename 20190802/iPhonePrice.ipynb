```python
from tqdm import tqdm_notebook

present_candi_text = []

# 검색된 문장을 저장
for n in tqdm_notebook(range(1, 1000, 10)):
    response = urlopen(html.format(num=n, key_word=urllib.parse.quote('아이폰 단통법')))

    soup = BeautifulSoup(response, "html.parser")

    tmp = soup.find_all('dl')

    for line in tmp:
        present_candi_text.append(line.text)
        
    time.sleep(0.5)
```

```python
# 저장했던 텍스트 출력
present_candi_text
```

```python
# 필요 패키지 import
import nltk
from konlpy.tag import Twitter; t = Twitter()
```

```python
# 변수에 문장 저장(단어 1만개)
present_text = ''

for each_line in present_candi_text[:10000]:
    present_text = present_text + each_line + '\n'
```

```python
# 단어 형태소 분석하여 출력
tokens_ko = t.morphs(present_text)
tokens_ko
```

```python
# 형태소 분석을 하고 나서 token을 가져온다
ko = nltk.Text(tokens_ko, name = '단통법')
print(len(ko.tokens))

# set으로 출현 단어 개수 출력
print(len(set(ko.tokens)))
```

```python
# 가장 많이 사용된 단어 출력
ko = nltk.Text(tokens_ko, name='단통법')
ko.vocab().most_common(100)
```

```python
# 필요 없는 단어 삭제
stop_words = ['.','가','요','답변','...','을','수','에','질문','제','를','이','도','한','할',
                      '좋','1','는','로','으로','2','것','은','다',',','니다','대','들',
                      '원금','들','데','..','의','때','겠','고','게','네요','한','일','할',
                      '10','?','하는','06','주','려고','인데','거','좀','는데','~','ㅎㅎ',
                      '하나','이상','20','뭐','까','있는','잘','습니다','다면','했','주려',
                      '지','있','못','후','중','줄','6','과','어떤','기본','!!',
                      '기','공','시','단', '통법', 'ㅇ','??',
                      '단어','선물해','라고','지','합','가요','....','보이','네','무지']

tokens_ko = [each_word for each_word in tokens_ko 
             if each_word not in stop_words]

# 삭제한 뒤에 제일 많이 나온 단어 순으로 출력
ko = nltk.Text(tokens_ko, name='여자 친구 선물')
ko.vocab().most_common(50)
```

```python
# 그래프로 단어와 빈도수 출력
plt.figure(figsize=(15,6))
ko.plot(50) 
plt.show()
```
