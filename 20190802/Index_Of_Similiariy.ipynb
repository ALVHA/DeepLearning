```python
## 문장의 유사도 측정
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# 최소 빈도 설정
vectorizer = CountVectorizer(1)
```

```python
## training 문장 넣기
contents = ko
```

```python
X = vectorizer.fit_transform(contents)
vectorizer.get_feature_names()
```

```python
contents_tokens = [t.morphs(row) for row in contents]
contents_tokens
```

```python
# 띄워져 있는 단어들을 공백을 사용하여 한 문장으로 길게 편집
contents_for_vectorize = []

for content in contents_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word
        
    contents_for_vectorize.append(sentence)
    
contents_for_vectorize
```

```python
# feature 찾기
X = vectorizer.fit_transform(contents_for_vectorize)
num_samples, num_features = X.shape
num_samples, num_features
```

```python
vectorizer.get_feature_names()
```

```python
X.toarray().transpose()
```

```python
# 새로운 문장을 학습시키고 위에서 했었던 함수 적용
new_post = ['이동통신단말장치에 관련한']
new_post_tokens = [t.morphs(row) for row in new_post]

new_post_for_vectorize = []

for content in new_post_tokens:
    sentence =''
    for word in content:
        sentence = sentence + ' ' + word
    
    new_post_for_vectorize.append(sentence)
    
new_post_for_vectorize
```

```python
new_post_vec = vectorizer.transform(new_post_for_vectorize)
# 새로운 문장을 벡터화
new_post_vec.toarray()
```

```python
# 새로운 문장과 원래 문장 사이의 거리를 구하는 함수
# 결과는 norm으로 일반화 시킨다.
import scipy as sp

def dist_raw(v1, v2):
    delta = v1 - v2
    return sp.linalg.norm(delta.toarray())
```

```python
best_doc = None
best_dist = 65535
best_i = None

for i in range(0, num_samples):
    post_vec = X.getrow(i)
    d = dist_raw(post_vec, new_post_vec)
    
    print("== Post %i with dist = %.2f : %s" %(i, d, contents[i]))
    
    if d<best_dist:
        best_dist = d
        best_i = i
```

```python
# 벡터화 시킨 것으로 문장 비교
for i in range(0, len(contents)):
    print(X.getrow(i).toarray())
    
print('---------------')
print(new_post_vec.toarray())
```

```python
# 각 벡터를 일반화시키고 거리를 구하는 함수를 만든다.
def dist_norm(v1, v2):
    v1_normalized = v1 / sp.linalg.norm(v1.toarray())
    v2_normalized = v2 / sp.linalg.norm(v2.toarray())
    
    delta = v1_normalized - v2_normalized
    
    return sp.linalg.norm(delta.toarray())
```

```python
# 거리를 구해보자
best_doc = None
best_dist = 65535
best_i = None

for i in range(0, num_samples):
    post_vec = X.getrow(i)
    d = dist_norm(post_vec, new_post_vec)
    
    print("== Post %i with dist=%.2f   : %s" %(i,d,contents[i]))
    
    if d<best_dist:
        best_dist = d
        best_i = i
```

```python
# 거리가 가장 가까운 문장 출력
print("Best post is %i, dist = %.2f" % (best_i, best_dist))
print('-->', new_post)
print('---->', contents[best_i])
```

```python
# tf = 단어의 빈도에 따른 중요도
# idf = 비교하는 문서에서 같은 단어가 있으면 핵심어휘일수 도있지만, 문서내에서는 아님
def tfidf(t, d, D):
    tf = float(d.count(t)) / sum(d.count(w) for w in set(d))
    idf = sp.log( float(len(D))/(len([doc for doc in D if t in doc])))
    return tf, idf
```

```python
# tfidf에 training set을 넣어준다
a, abb, abc = ['a'], ['a', 'b', 'b'], ['a', 'b', 'c']
D = [a,abb, abc]

# a가 모든 set에 나왔기 때문에 1이라고 출력
print(tfidf('a', a, D))
print(tfidf('b', abb, D))
print(tfidf('a', abc, D))
print(tfidf('b', abc, D))
```

```python
# 필요 패키지 import
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(decode_error='ignore')
```

```python
# 단어 형태소 분석
contents_tokens = [t.morphs(row) for row in contents]

contents_for_vectorize = []

# 띄어쓰기로 구분된 단어를 공백을 사용하여 길게 한줄로 출력
for content in contents_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word
        
    contents_for_vectorize.append(sentence)
    
# 단어를 벡터화
X = vectorizer.fit_transform(contents_for_vectorize)
num_samples, num_features = X.shape
num_samples, num_features
```

```python
# 새로운 문장 학습
new_post = ['이동통신을 보는 단통법']
new_post_tokens = [t.morphs(row) for row in new_post]

new_post_for_vectorize = []

# 띄어쓰기로 나누어진 단어를 공백을 사용해 한 줄로 만들기
for content in new_post_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word
        
    new_post_for_vectorize.append(sentence)
    
new_post_for_vectorize
```

```python
best_doc = None
best_dist = 65535
best_i = None

# 벡터화시킨 문장을 원래 문장과의 거리를 비교하여 거리를 계산하고 거리가 가까운 문장을 출력한다.
for i in range(0, num_samples):
    post_vec = X.getrow(i)
    d = dist_norm(post_vec, new_post_vec)
    
    print("== Post %i with dist=%.2f   : %s" %(i,d,contents[i]))
    
    if d<best_dist:
        best_dist = d
        best_i = i
        
print("Best post is %i, dist = %.2f" % (best_i, best_dist))
print('-->', new_post)
print('---->', contents[best_i])
```
