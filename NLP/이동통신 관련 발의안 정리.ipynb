```python
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image

# kobill 모듈 import
from konlpy.corpus import kobill
files_ko = kobill.fileids()

import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import font_manager, rc
import platform

path = "c:/Windows/Fonts/malgun.ttf"
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
# 텍스트 파일 읽기, 단통법 처리된 의안
doc_ko = open('data/단통법.txt').read() ## kobill 은 국회 의안 말뭉치
```

```python
doc_ko
```

```python
# Twitter 모듈 import
from konlpy.tag import Twitter;
t = Twitter()

# 명사만 출력
tokens_ko = t.nouns(doc_ko)
tokens_ko
```

```python
#텍스트 이름 수정하기
ko = nltk.Text(tokens_ko, name = '대한민국 국회 단통법 발의안')
```

```python
# 토큰의 숫자, 최초 출현 토큰 숫자 출력
print(len(ko.tokens))           
print(len(set(ko.tokens)))   
ko.vocab()  
```

```python
# 단어 빈도 출력
plt.figure(figsize = (12,6))
ko.plot(50)
plt.show()
```

```python
# 불용어 ( 의미  없는 문자) 설정
stop_words = ['생', '략','.','제','항', '(', ')', ',', "'", '%', '-', 'X', ').', '×','의','자','에','안','번', '호','을','이','다','만','로','가','를']

# 불용어를 제거한 단어들 설정 및 출력
ko = [each_word for each_word in ko if each_word not in stop_words]
ko
```

```python
ko = nltk.Text(ko, name='대한민국 국회 단통법 발의안')

# 위에서 정제한 데이터 출력
plt.figure(figsize=(12,6))
ko.plot(50)     
plt.show()
```

```python
# 특정 단어가 얼마나 많이 나왔는지 확인하기.
for i in range(len(ko)):
    a = str(ko.count(ko[i]))
    print(ko[i] + ':' + a)
```

```python
# 특정 단어들이 어디에 몇번이나 출현했는지 알아보는 그래프
plt.figure(figsize=(12,6))
ko.dispersion_plot(['이동통신', '사업자', '유통'])
```

```python
# '이동통신'와 같이 나온 주변 단어들 출력
ko.concordance('이동통신')
```

```python
# 연관어 출력
ko.collocations()
```

```python
data = ko.vocab().most_common(150)

# c:/Windows/Fonts/malgun.ttf
# 위에서 정제한 단어들을 그림으로 출력
wordcloud = WordCloud(font_path='/Library/Fonts/malgun.ttf', 
                      relative_scaling = 0.2, 
                      background_color='white').generate_from_frequencies(dict(data))

plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
